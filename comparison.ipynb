# Quick and easy data visualization in Python
Comparison of the standard visualization between the packages:
- Matplotlib (pandas.plot)
- Seaborn
- Plotly-Express

```python
# import data
import plotly.express as px

df = px.data.gapminder().query("year == 2007")
df.head()
```

## Show the population for each continent/country

```python
# matplot (pandas default)
df.plot.bar(y="pop", x="country", title="Title")
```

```python
# seaborn
import seaborn as sns
sns.barplot(data=df, y="pop", x="country").set_title("title")
```

```python
# plotly
px.bar(df, y="pop", x="country", title="Title")
```

## Show the relationship between life expectancy and income

```python
# matplot (pandas default)
df.plot.scatter(x="gdpPercap", y="lifeExp")
```

```python
# seaborn
sns.scatterplot(data=df, x="gdpPercap", y="lifeExp", size="pop", sizes=(20,500), hue="continent")
```

```python
# plotly
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent")
fig.show()
fig.write_html("interactiveplot.html")
```

# Styling Plotly even further

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x="gdpPercap", y=np.sin(t),
    name='sin',
    mode='markers',
    marker_color='rgba(152, 0, 0, .8)'
))

fig.add_trace(go.Scatter(
    x=t, y=np.cos(t),
    name='cos',
    marker_color='rgba(255, 182, 193, .9)'
))

# Set options common to all traces with fig.update_traces
fig.update_traces(mode='markers', marker_line_width=2, marker_size=10)
fig.update_layout(title='Styled Scatter',
                  yaxis_zeroline=False, xaxis_zeroline=False)


fig.show()
